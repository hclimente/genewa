```r
library(tidyverse)
library(magrittr)
```

```r
new <- read_tsv('snp2hgnc.tsv', col_types = 'cc') %>%
    mutate(new = TRUE)
old <- read_tsv('../genesis/gene2snp.hg19', col_types = 'cc') %>%
    set_colnames(c('snp', 'symbol')) %>%
    mutate(old = TRUE)

snp2gene <- full_join(old, new, by = c('snp', 'symbol')) %>%
    mutate(old = !is.na(old), new = !is.na(new))
```

```r
select(snp2gene, new, old) %>% table
```

```r
old_genes <- snp2gene %>% filter(old & ! new) %>% .$symbol %>% unique
length(old_genes)

sum(old_genes %in% new$symbol)
```

```r
old_genes[! old_genes %in% new$symbol]
```

```r
hgnc <- read_tsv('non_alt_loci_set.txt')
filter(hgnc, symbol %in% old_genes[! old_genes %in% new$symbol]) %>% nrow
filter(hgnc, symbol %in% old_genes[! old_genes %in% new$symbol] & !is.na(ensembl_gene_id)) %>% nrow
```

```r
hgnc %>%
    filter(symbol %in% old_genes[! old_genes %in% new$symbol] & !is.na(ensembl_gene_id)) %>% 
    mutate(chr = strsplit(location, '[pq]') %>% lapply(function(x) { x[1]}) %>% unlist) %>%
    .$chr %>%
    table
```

```r
new_genes <- snp2gene %>% filter(old & ! new) %>% .$symbol %>% unique
length(new_genes)

sum(new_genes %in% old$symbol)
```
